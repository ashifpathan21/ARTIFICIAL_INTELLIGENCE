**Numpy indexing and slicing of vectors**

```python
import numpy as np
```

```python
arr = np.arange(11,21)
arr
```

```python
arr[4]
```

```python
arr[1:4]
```

```python
arr[1:10:2]
```

***Numpy Indexing & Slicing of Matrix***

```python
arr = np.arange(1,31).reshape(6,5)
```

```python
arr
```

```python
arr[0]
```

```python
arr[0][4]
arr[0,4]
```

```python
arr[0:3]
```

```python
arr[0:3]
```

```python
arr[0:2 , 1:3]
```

```python
arr
```

```python
arr[3: , 3:]
```

**Boolean Indexing**

```python
bool_index= arr%2 == 0
bool_index
```

```python
arr = arr[bool_index]
```
