**Deeep & Shallow Copy**

```python
ar1 = np.arange(1,21)
```

```python
ar2 = ar1[:5]
```

```python
ar2
```

```python
ar2=ar2+10
```

```python
ar1 # no change due to shallow copy
```

```python
b = ar1.copy()
```

```python
b
```

```python
b = ar1 # deep copy
```

**Multiplication of Matrix**

```python
A = np.arange(1,7).reshape(2,3)
B = np.arange(7,13).reshape(3,2)
```

```python
A @ B # according to maths
```

```python
# also
np.dot(A , B)
```

```python
# transpose of array - row become column
A.T
```

***Advance array manipulation***

```python
#! Stacking of array
a = np.arange(1,5)
b = np.arange(6,10)
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```
