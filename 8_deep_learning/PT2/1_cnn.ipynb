```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Conv2D , Flatten , MaxPooling2D ,Dropout
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv('mnist_train.csv')
df_test = pd.read_csv('mnist_test.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
X_train = df.drop(columns=['label'] , axis=1).values
y_train = df['label']
X_test = df_test.drop(columns=['label'] , axis=1 ).values
y_test = df_test['label']
```

```python
X_train = X_train.astype('Float32')/255.0
X_test = X_test.astype('Float32')/255.0
```

```python
X_train_img= X_train.reshape(-1 , 28,28)
X_test_img = X_test.reshape(-1 , 28,28)
```

```python
y_train_cat = to_categorical(y_train , 10)
y_test_cat = to_categorical(y_test , 10)
```

```python
per = Sequential([
  Flatten(input_shape=(28,28)),
  Dense(10, activation='softmax')
])
```

```python
per.compile(optimizer='sgd' , loss='categorical_crossentropy' , metrics=['accuracy'])
```

```python
hist_per = per.fit(X_train_img , y_train_cat,epochs=5,batch_size=32 , validation_data=(X_test_img,y_test_cat) ,verbose=1)
```

```python
acc_per = per.evaluate(X_test_img , y_test_cat , verbose=0)[1]
```

```python
print(acc_per)
```

```python
# ann 
ann = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128 , activation='relu'),
    Dense(64 , activation='relu'),
    Dense(10 , activation='softmax')
])
```

```python
ann.compile(optimizer='adam' , loss='categorical_crossentropy' , metrics=['accuracy'])
```

```python
hist_ann = ann.fit(X_train_img , y_train_cat,epochs=5,batch_size=32 , validation_data=(X_test_img,y_test_cat) ,verbose=1)
```

```python
acc_ann = ann.evaluate(X_test_img , y_test_cat , verbose=0)[1]
acc_ann
```

```python
X_train_img= X_train.reshape(-1 , 28,28,1)
X_test_img = X_test.reshape(-1 , 28,28,1)
```

```python
#cnn
cnn = Sequential([
    Conv2D(32 , kernel_size=(3,3) , activation='relu' , input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64, kernel_size=(3,3) , activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(128,activation='relu'),
    Dropout(0.5),
    Dense(10,activation='softmax')
])
```

```python
cnn.compile(optimizer='adam' , loss='categorical_crossentropy' , metrics=['accuracy'])
```

```python
hist_cnn = cnn.fit(X_train_img , y_train_cat,epochs=5,batch_size=32 , validation_data=(X_test_img,y_test_cat) ,verbose=1)
```

```python
acc_cnn = cnn.evaluate(X_test_img , y_test_cat , verbose=0)[1]
acc_cnn
```
