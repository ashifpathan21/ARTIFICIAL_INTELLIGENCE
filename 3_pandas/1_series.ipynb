***Series***

one dimensional laballed array capable of holding any type of data the axis labels are collectively called index


```python
import numpy as np
import pandas as pd 
```

```python
arr = np.array([1, 3, 4 ])
labels = ['a' , 'b'  ,'c']
my_list = [1,2,3]
dic = {1:1,3:2,4:5}
```
