Finding Missing Data

```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

removing nan

```python
df.dropna()
```

```python
#at least thresh non null value
df.dropna(thresh=3)
```

```python
df.fillna(0)
```

```python
values = {"A": 0 , "B":10 , "C":12 , "D":11} 
```

```python
df.fillna(value=values)
```
