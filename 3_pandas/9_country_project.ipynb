```python
# which country has the highest population
df[df["population"] == df["population"].max()]["country"]
```

```python
# what is the capital of the highest population country
df[df["population"] == df["population"].max()]['capital_city']
```

```python
# capital of the country with least population 
df[df["population"] == df["population"].min()]["country"]
df[df["population"] == df["population"].min()]["capital_city"]
```

```python
# top 5 countries with highest democratic score 
df.columns
```

```python
# top 5 countries with highest democratic score 
df.sort_values(by="democracy_score" , ascending=False).head()
```

```python
# how many total regions are there
df["region"].value_counts()
```

```python
# countries lie in Eastern Europe region
df[df["region"] == "Eastern Europe"]
```

```python
# who is the leader of 2nd highest populated country
df[df["population"] == df["population"].nlargest(2).iloc[1]]["political_leader"]
```

```python
# how many countries are there whose political leaders are unknown
df[df["political_leader"].isna()]["country"]
```

```python
# how many country has Republic in there full name 
df[df["country_long"].apply(lambda txt : True if "Republic" in txt else False)]['country'].count()
```

```python
africadf = df[df["continent"] == "Africa"]
africadf[africadf["population"] == africadf["population"].max()]
```
