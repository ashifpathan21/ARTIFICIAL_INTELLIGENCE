Creating a Dataframe

```python
import pandas as pd 
import numpy as np
```

```python
data  = {
    "Name":["Ram" , "Shyam" , "Sita"],
    "Age":[20,21,19],
    "City":["Agra" , "Delhi" , "Agra"],
    "Salary":[20000,25000,20000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df 
```
