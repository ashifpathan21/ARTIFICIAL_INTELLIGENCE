```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2 
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

**Subplots**

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
```

**Object oriented way**

```python
fig = plt.figure(figsize=(8,6) , dpi=100)
```

```python
#[marginleft , marginbottom , width , height] value btw 0 and 1
ax1 = fig.add_axes([0.1,0.1,0.8,1])
ax1.plot(x,y)
ax2 = fig.add_axes([0.4 , 0.4 , 0.5 , 0.5])
ax2.plot(y,x)
```
