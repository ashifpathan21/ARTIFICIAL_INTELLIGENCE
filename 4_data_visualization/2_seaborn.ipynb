**Distribution Plot**

for Quantitative Variable - (1,10,100,23,42 ....)
* histplot/distplot
* jointplot
* pairplot
* rudplot

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

distplot/histplot , displot

```python

sns.displot(df["total_bill"])
sns.displot(df["total_bill"],bins=3)
```

```python

sns.histplot(df["total_bill"] , kde=True)
```

```python
plt.subplot(1,2 , 1)
sns.histplot(df["total_bill"] , kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"] , kde=True)
```

jointplot

```python
sns.jointplot( x="total_bill" , y="tip" , data=df  , kind="scatter")
```

```python
sns.jointplot( x="total_bill" , y="tip" , data=df  , kind="hex")
```

```python
sns.jointplot( x="total_bill" , y="tip" , data=df  , kind="kde")
```

```python
sns.jointplot( x="total_bill" , y="tip" , data=df  , kind="reg")
```

```python
sns.jointplot( x="total_bill" , y="tip" , data=df  , kind="resid")
```

pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df , hue="sex")
```

```python
sns.pairplot(df , hue="size" , palette="rainbow")
```

rugplot

```python
sns.rugplot(df)
```

* stripplot
* swarmplot

```python
sns.countplot(x=df["sex"] , hue=df["smoker"])
```

```python
sns.barplot(x=df["sex"] , y=df["total_bill"])
```

```python
sns.boxplot(x="tip" , y="day" , data=df)
```

```python
sns.violinplot(x="total_bill" , y="day" , data=df)
```

```python
sns.stripplot(x="day" , y="total_bill" , data=df)
```

**MatrixPlot**


Heat map 

```python
flights  = sns.load_dataset("flights")
flights
```

```python
tips = df[["total_bill" , "size" , "tip"]]
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr() , annot=True)
```

Cluster map

```python
sns.clustermap(tips.corr() )
```

Pivot table heat map


```python
flights_piv = flights.pivot_table(values="passengers" , index="month" , columns="year")
```

```python
flights_piv
```

```python
sns.heatmap(flights_piv)
```
