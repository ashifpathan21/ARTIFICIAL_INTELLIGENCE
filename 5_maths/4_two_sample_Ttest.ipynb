```python
import numpy as np 
from scipy.stats import t 

```

```python
sample_1 = [1,3,4,5,6,8,9]
s1_size = len(sample_1)
s1_mean = np.mean(sample_1)
s1_std = np.std(sample_1)
sample_2 = [5,3,9,6,8,12,10]
s2_size = len(sample_2)
s2_mean = np.mean(sample_2)
s2_std = np.std(sample_2)

```

```python
t_stats = (s1_mean - s2_mean)/np.sqrt((s1_std**2 / s1_size)+ (s2_std**2 / s2_size))
```

```python
t_stats
```

```python
p_stats = 2*(1-t.cdf(abs(t_stats),df=(min(s1_size,s2_size)-1)))
p_stats
```

```python
alpha = 0.05 
def check(p_stats , alpha):
    if p_stats < alpha:
        print("Reject the null hypothesis")
    else:
        print("Accept the null hypothesis")

check(p_stats,alpha)
```

```python
from scipy import stats 
t_val , p_val =  stats.ttest_ind(sample_1 , sample_2 , equal_var=False)
print(t_val , p_val)
print(t_val == t_stats)
print(p_val == p_stats)
check(p_val  , alpha)
```
