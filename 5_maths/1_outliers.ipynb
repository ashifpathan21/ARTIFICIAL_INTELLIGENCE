```python
import numpy as np
arr = np.array([2,3,4,5,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
```

```python
# 5 number summary
min = arr.min()
q1 = np.percentile(arr,25)
q2 = np.percentile(arr , 50)
q3 = np.percentile(arr,75)
max = arr.max()
print(min , q1 , q2 , q3 , max)
```

```python
range = max - min
IQR = q3-q1
print(range , IQR)
```

```python
lower_fence = q1 - (1.5*IQR)
upper_fence = q3 + (1.5 * IQR)
print(lower_fence , upper_fence)
```

```python
# removing outliers
arr = arr[(arr >= lower_fence) & (arr <= upper_fence)]
```

```python
import seaborn as sns
sns.boxplot(x=arr)
```
